```python
import sklearn
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(10000000)
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

sns.distplot(s)
```

```python
np.random.seed(10000000)
x = stats.norm.rvs(loc = 0, scale = 3, size = 100)
sns.distplot(x)
plt.show()
```

always want to check if we have a normal distribution

```python
stats.shapiro(x)
#teststatistic, p-value ( > 0.05 to be normal )
```

```python
1st is test stat Z-test stat
2nd is p-value
```

```python
x_phack = x**2
stats.shapiro(x_phack)
# see how just squaring data drive p-val super low
```

```python
y = stats.norm.rvs(loc=5, scale = 3, size=100)
stats.shapiro(y)
```

```python
stats.f_oneway(x,y)
###F_onewayResult(statistic=131.46924952585854, pvalue=1.1363782157819309e-23)
#because p value is so small, they are very very different, not the same
# less than 0.05 for p value means two distributions are significantly different
# above 0.5 is normal distribution for 1 distribution
```
